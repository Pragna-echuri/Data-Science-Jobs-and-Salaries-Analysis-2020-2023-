```python
job_title_counts = df['job_title'].value_counts()
top_10_job_titles = job_title_counts[:10].index
df_top_10_job_titles = df[df['job_title'].isin(top_10_job_titles)]

loc_count = df['company_location'].value_counts()
top_10__loc = loc_count[:10].index
df_top_10_loc = df[df['company_location'].isin(top_10__loc)]

cross_table = pd.crosstab(df_top_10_loc['company_location'], df_top_10_loc['experience_level'])

plt.figure(figsize=(8, 6))
sns.heatmap(cross_table, cmap='Blues', linewidths=0.5, annot=True, fmt='d')

plt.title('Frequency Distribution of Company Sizes Across Top 10 Locations')
plt.xlabel('Company Size')
plt.ylabel('Company Location')

plt.show()
```

```python
grouped_data = data2.groupby(['work_year', 'job_category']).size().unstack()
grouped_data.plot(kind='line',marker='>')
plt.title('Year wise Job Category growth')
plt.xlabel('Year-->')
plt.ylabel('Number of jobs-->')
plt.legend(title='Job Category', bbox_to_anchor=(0.5, 1),fontsize=8, loc='upper right')
plt.grid(False)
#plt.show()
plt.savefig('ne5.png')
```

```python
top_10_locations = data2['company_location'].value_counts().nlargest(10).index

# Filter the data to include only the top 10 locations
df_filtered = data2[data2['company_location'].isin(top_10_locations)]

# Group the filtered data by 'job_category' and 'company_location' and calculate the count of occurrences
grouped_data = df_filtered.groupby(['job_category', 'company_location']).size().unstack()

# Plot the grouped bar chart
plt.figure(figsize=(50,30))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Job Category Distribution by Top 10 Company Locations')
plt.xlabel('Job Category')
plt.ylabel('no of employees')
plt.xticks(rotation=30,ha='right')
plt.legend(title='company location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('new6.png')
```
